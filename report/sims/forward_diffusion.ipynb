```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_trials = 100
T = 100
beta = 0.5
x0 = 10

xs = np.empty((n_trials, T))
xs[0:,] = x0

for i in range(n_trials):
    xt = x0
    for t in range(1, T):
        xt = np.sqrt(1-beta) * xt + np.sqrt(beta)*np.random.randn()
        xs[i, t] = xt
```

```python
plt.plot(xs.T, 'b.', alpha=0.4)
plt.xlabel("Time-step [t]")
plt.ylabel("X at time t")
plt.title("One-dimensional forward diffusion process")
plt.savefig("../figures/simple_diffusion.png", transparent=True)
```
